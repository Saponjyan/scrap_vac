```python
# get links to all vacancies and add them to the list
soup=BeautifulSoup(driver.page_source,"lxml")
link=soup.find_all("a",class_="base-card__full-link")
links=[]
for l in link:
    links.append(l.get("href")+"&_l=en_US") # translate into english
```

```python
# check number of links
len(links)
```

```python
# function for getting information from the link to the offer by sending a request, converting the response into a soup object and processing it with bs methods

headers=({'user-agent': 
          'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36'
           '(KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36'})


def scan(url):

    response = requests.get(url,headers=headers)
    sleep(0.5)                  # when time decreases, more links give no information
    soup = BeautifulSoup(response.text,"lxml")

    # company name
    try:
        company = " ".join(soup.find("a",
              class_="topcard__org-name-link topcard__flavor--black-link").text.split())
    except:
        company = "unknown"


    # title    
    try:
        title = soup.find("h1",
        class_="top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title").text
    except:
        title = "unknown"

    try:
        xxx = soup.find_all("ul",class_="description__job-criteria-list")
        for ii in xxx:
            y = " ".join(ii.text.split())
    except:
        pass
    
    
    # field of company
    try:
        ind = " ".join(y.split('Industries')[1].split())
        y = y.split('Industries')[0]
    except:
        ind = "unknown"


    # employment type
    try:
        emp = " ".join(y.split('Employment type')[1].split())
        y = y.split('Employment type')[0]
    except:
        emp = "unknown"


    # seniority level
    try:
        level = " ".join(y.split('Seniority level')[1].split())
    except:
        level="unknown"

    # location
    try:
        loc = " ".join(soup.find("span",class_="topcard__flavor topcard__flavor--bullet").text.split())
    except:
        loc = "unknown"
    
    # publication time
    try:
        time = " ".join(soup.find("span",
            class_="posted-time-ago__text posted-time-ago__text--new topcard__flavor--metadata").text.split())
    except:
        time = "unknown"

    # link
    link = url


    # description
    try:
        card_text = " ".join(soup.find("div",
                class_="show-more-less-html__markup show-more-less-html__markup--clamp-after-5").text.split())
    except:
        card_text = "unknown"

    # country
    try:
        country = loc.split(',')[-1]
    except:
        country = "unknown"


        
    return ([title, company, country, level, emp, ind,loc,time,link,card_text])
```

```python
# scanning all links
%%time
data = []
for i in tqdm(links):
    data.append(scan(i)) 
```

```python
driver.quit()
```

```python
# creating datаframe
df=pd.DataFrame(data,columns=['title','company','country','level',
                              'employment_type','company_field',
                              'location','date','link','description'])
```

```python
now = datetime.now()

df['date'] = df['date'].replace({' hours ago': '', ' hour ago': ''}, regex=True)
df['date'] = df['date'].str.replace('.* ago', '0', regex=True)  
df['date'] = df['date'].apply(lambda x: x if x== 'unknown' else (now - timedelta(hours=int(x))).date())      
```
